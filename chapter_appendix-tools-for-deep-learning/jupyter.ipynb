Importar bibliotecas

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

Fazer requisição da página

```python
url = "https://pt.wikipedia.org/wiki/Bad_Bunny"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    print("Página acessada com sucesso!")
else:
    print("Erro ao acessar a página")
```

Criar o objeto BeautifulSoup

```python
soup = BeautifulSoup(response.content, "lxml")
```

Extrair informações principais
Titulo

```python
titulo = soup.find("h1").text
print(titulo)
```

Primeiro paragrafo

```python
primeiro_paragrafo = soup.find("p").text
print(primeiro_paragrafo)
```

Tabela lateral

```python
infobox = soup.find("table", class_="infobox")

dados_infobox = {}

for linha in infobox.find_all("tr"):
    if linha.th and linha.td:
        chave = linha.th.text.strip()
        valor = linha.td.text.strip()
        dados_infobox[chave] = valor

print(dados_infobox)
```

Extrair links internos da página

```python
links = []

for link in soup.find_all("a"):
    href = link.get("href")
    if href and href.startswith("/wiki/") and ":" not in href:
        links.append("https://pt.wikipedia.org" + href)

links_unicos = list(set(links))

print(f"Total de links internos: {len(links_unicos)}")
print(links_unicos[:10])
```

Salvar os dados

```python
df_links = pd.DataFrame(links_unicos, columns=["Link"])
df_links.to_csv("links_bad_bunny.csv", index=False)
```

```python
with open("infobox_bad_bunny.json", "w", encoding="utf-8") as f:
    json.dump(dados_infobox, f, ensure_ascii=False, indent=4)
```
